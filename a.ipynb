```python
from untrade.client import Client

def perform_backtest(csv_file_path):
    """
    Performs a backtest using the Untrade SDK on the specified CSV file.
    
    Parameters:
        csv_file_path (str): Path to the CSV file with strategy results.
    
    Returns:
        result: Backtest results.
    """
    client = Client()
    try:
        print("Starting backtest...")
        result = client.backtest(
            file_path=csv_file_path,
            leverage=1,
            jupyter_id="screening",
        )
        print("Backtest completed.")
        if result is None:
            print("Error: The backtest result is None. Please check the input file and configuration.")
        print("returning result")
        return result
    except Exception as e:
        print(f"Error during backtest: {e}")
        return None

if __name__ == "__main__":
    # Path to the CSV file generated by the strategy
    print("loading csv file")
    csv_file_path = "/Users/shivanshgupta/Desktop/zelta hack/singals_BTC_2019_2023_15m.csv"
    print("csv file loaded")
    print("performing backtest")
    backtest_result = perform_backtest(csv_file_path)
    print("backtest performed")
    
    if backtest_result is None:
        print("No backtest result to display.")
    else:
        print("Backtest result:")
        for value in backtest_result:
            print(value)
```
